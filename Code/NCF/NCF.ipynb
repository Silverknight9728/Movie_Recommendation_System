```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
from data_process import clean_data 
from data_dummy import data_split
from model import create_model
from eval import eval_matrix
```

```python
movie_path = './ml-latest-small/movies.csv'
user_rat_path = './ml-latest-small/ratings.csv'
df_movies = pd.read_csv(movie_path)
df_rating  = pd.read_csv(user_rat_path)
```

```python
df_rating
```

```python
df_movies,df_rating = clean_data(df_movies, df_rating)
```

```python
df_rating
```

```python
x, model_train, model_test = data_split(df_rating)
```

```python

NCF = create_model(df_rating, 50,"adam",'mean_absolute_error')
```

```python
# train
epochs = 50
batch_size = 128
split = 0.2
NCF = NCF.fit(x, y=model_train["y"], 
                     epochs=epochs, batch_size = batch_size,shuffle=True, verbose=1, validation_split=split)
```

```python
model_test["yhat"] = NCF.model.predict([model_test["user"], model_test["product"]])
```

```python
precision, recall, ndcg,f1 = eval_matrix(model_test)
```
